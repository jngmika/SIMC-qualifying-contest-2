## Task 1

###  Build a criterion that, with the best possible accuracy, differentiates lizards of species #5 from all other lizards and uses only the femoral pore number on the right side (FPNr).
### *Hint: plot and explore the FPNr distribution of the lizards with regard to their species.*

```python
df = pd.read_csv("lizards_data.csv")
```

```python
df[["Species_num", "FPNr"]]
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter(df["Species_num"], df["FPNr"])
ax.set_xlabel("Species_num")
ax.set_ylabel("FPNr")
ax.set_yticks([2 * i for i in range(1, 10)])
```

as we can assume from the graph only lizards from the species group number 5 have FPNr < 12 (except for the outlier from the group number 4)

```python
df.loc[df["FPNr"] < 12]["Species_num"]
```

```python
df.loc[df["FPNr"] < 12].loc[df["Species_num"] == 5]
```

```python
df.loc[df["FPNr"] < 12].loc[df["Species_num"] == 4]
```

there is only one outlier (group #4 falsely classified as group #5)

## Task 2

### Build a criterion that, with the best possible accuracy, differentiates lizards of species #5 from all other lizards and uses two variables out of the measured morphometric and pholidosis characteristics.
### *Hint: one of the methods to find the best pair of predicting variables (predictors) may be exhaustive search over all possible pairs of variables.*

```python
df
```

as we saw in the previous task criterion **FPNr < 12** is quite good to distinguish species #5 from the other groups except for the #4. let's find the best one to distinguish group #5 from group #4 (to distinguish the outliers from the distinguishing by FPNr)

## answer: FPNr < 12

```python
df2 = df.loc[(df["Species_num"] == 4) | (df["Species_num"] == 5)]
df2
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter(df2["Species_num"], df2["MBS"])
ax.set_xlabel("Species_num")
ax.set_ylabel("MBS")
ax.set_xticks([4, 5])
ax.set_yticks([40 + i for i in range(5)])
```

by plotting MBS and Species_num we can clearly see that MBS > 44 for every lizard from group #4 and MBS < 42 for every lizard from group #5

thus, we can distinguish group #5 by these criteria:

**FPNr < 12** ans **MBS < 42**

## answer: FPNr < 12 & MBS < 42

```python
df.loc[(df["FPNr"] < 12) & (df["MBS"] < 42)]
```

```python
df.loc[(df["FPNr"] < 12) & (df["MBS"] < 42)]["Species_num"].unique()
```

#### with this metric we can find lizards from group #5 with no outliers

## Task 3
### Build a criterion that, with the best possible accuracy, predicts lizards’ sex regardless of their biological species on the basis of the morphometric and/or pholidosis characteristics.
### *Hint from the biologists: it is expected (but not guaranteed!) that sex is correlated with the ratios of some measured linear sizes; but this does preclude from using other predictors in the criterion.*

```python
df
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter(df["HL"] / (df["SVL"] + df["TRL"]), df["Sex"])
ax.set_xlabel("HL / (SVL + TRL)")
ax.set_ylabel("Sex")
```

as we can see HL / (SVL + TRL) distinguishes female and male lizards.

## answer: HL / (SVL + TRL) > 0.21 - male, others - female
